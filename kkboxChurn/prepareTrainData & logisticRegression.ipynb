```python
transactionDf = pd.concat([pd.read_csv('data/transactions.csv') , pd.read_csv('data/train_v2.csv')])
# transactionDf= pd.read_csv('data/transactions.csv')
transactionDf.head()
```

```python
# transactionDf.msno.unique().shape
transactionDf = transactionDf.loc[(transactionDf['msno'].isin(uniqueUsers)),['msno','is_auto_renew','is_cancel','membership_expire_date', 'transaction_date']]
transactionDf.head()
```

```python
transactionDf.membership_expire_date = transactionDf.membership_expire_date.fillna(method='ffill')
transactionDf.transaction_date = transactionDf.transaction_date.fillna(method='ffill')
transactionDf.membership_expire_date  = transactionDf.membership_expire_date.astype(int)
transactionDf.transaction_date = transactionDf.transaction_date.astype(int)
expire_year = transactionDf.membership_expire_date//10000
expire_month = (transactionDf.membership_expire_date%10000)//100
expire_day = transactionDf.membership_expire_date//100
tran_year= transactionDf.transaction_date//10000
tran_month = (transactionDf.transaction_date %10000)//100
tran_day = transactionDf.transaction_date//100
# print expire_year
transactionDf['daysBefore'] = (expire_year - tran_year) * 365 + (expire_month - tran_month) * 30 + (expire_day - tran_day)
transactionDf.head()
```

```python
transactionDf = transactionDf.loc[:,['msno','is_auto_renew','is_cancel','daysBefore']].groupby('msno').agg('mean').reset_index()
transactionDf.loc[transactionDf.is_cancel >0,'is_cancel'] = 1
transactionDf.head()
```

```python
trainUsersDf = trainDf.join(transactionDf.set_index('msno'), on='msno')
trainUsersDf.head()
```

```python
trainUsersDf = trainUsersDf.dropna()
trainUsersDf.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.ensemble import GradientBoostingClassifier
X = trainUsersDf[['is_auto_renew','is_cancel','daysBefore']]
y = trainUsersDf.is_churn
clf= GradientBoostingClassifier(max_depth=20)
clf.fit(X,y)
predicted = clf.predict(X)
accuracy_score(predicted,y)
```
